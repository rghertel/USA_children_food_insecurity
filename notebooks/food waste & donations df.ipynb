```python
import re
import pandas as pd
```

```python
#import csv

food_waste = pd.read_csv('data/ReFED_US_State_Food_Surplus_Detail.csv', low_memory=False)

food_waste.head()
```

```python
#drop two NAN rows and turn third row into column names

food_waste = food_waste.drop([0,1])
food_waste.columns = food_waste.iloc[0]
food_waste
```

```python
#drop row with column names, and only keep important rows

food_waste = food_waste.drop([2])
food_waste = food_waste[['year','state','sector','tons_surplus','us_dollars_surplus','tons_waste','tons_donated']]
food_waste
```

```python
#filter to 2018 only

food_waste_18 = food_waste[food_waste['year'] == '2018']
food_waste_18.info()
```

```python
#change datatypes hosp_stay = hosp_stay.astype({'Attending Provider NPI':object})

food_waste_18= food_waste_18.astype({'tons_surplus':float,'us_dollars_surplus':float,'tons_waste':float, 'tons_donated':float})
```

```python
food_waste_18.info()
```

```python
#group by state, sum the other three columns

food_waste_18 = food_waste_18.groupby('state').sum().reset_index()
food_waste_18.head()
```

```python
food_waste_18['dollar_per_ton'] = food_waste_18['us_dollars_surplus']/food_waste_18['tons_surplus']
```

```python
food_waste_18['us_dollars_waste'] = food_waste_18['tons_waste']*food_waste_18['dollar_per_ton']
```

```python
food_waste_18['us_dollars_donated'] = food_waste_18['tons_donated']*food_waste_18['dollar_per_ton']
```

```python
food_waste_18.to_csv('data/food_waste_18.csv')
```
