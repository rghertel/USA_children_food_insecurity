```python
main_URLs = []
initials = ['AL','AK','AZ','AR','CA','CO','CT', 'DC','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN',
            'MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA',
            'WA','WV','WI','WY']

for i in range(len(initials)):
    address = URL + 'state/' + initials[i] + '.html'
    main_URLs.append(address)
    
main_URLs

# for address in states URLS:
#     if state URL has multiple pages, grab those pages, and append them to state URL list
```

```python
#create a new empty list for all main pages, and all additional pages. Iterate through main pages twice, first to add mains to
#new list, then add additionals to new list.

all_URLs = []
    
for i in main_URLs:
    all_URLs.append(i)

    
for e in main_URLs:
    state = requests.get(e)
    ssoup = BeautifulSoup(state.content, 'lxml')
    extra_pages = ssoup.find_all('a', href=re.compile('index'))
    
    for i in extra_pages:
        if i != extra_pages[-1]:
            url_tail = i['href']
            #print(url_tail)
            full_address = URL + 'state/' + url_tail
            #print(full_address)
            all_URLs.append(full_address)
        else:
            #url_tail=False
            #print("don't need last one")
            pass
        
print(all_URLs)
```

```python
#create a for loop to grab the table tag from pages; create empty df, create soup, extract tables, concatenate

fb_df = pd.DataFrame()

for info in all_URLs:
    page = requests.get(info)
    soup = BeautifulSoup(page.content, 'lxml')
    tables = soup.find_all('table', class_='default')
    result_list = pd.read_html(str(tables))[0]
    fb_df = pd.concat([fb_df, result_list])
    
fb_df
```

```python
#drop irrelevant columns

fb_df = fb_df.drop(columns=['Details','Phone'])
fb_df
```

```python
#create a column for state; for loop to iterate over address column, extract state

state = []

for i in fb_df['Address']:
    state_code = re.findall('[A-Z]{2}',i) #this creates a list, not a string
    state.append(state_code)

#add list as a new column

fb_df['State'] = state


#some street abbreviations were extracted in above for loop; remove first element if there are two
for states in fb_df['State']:
    if len(states) == 2:
        states.remove(states[0])
    else:
        pass
    #str(states) #i want to cast as string, but not working right now

fb_df.head(20)
```
